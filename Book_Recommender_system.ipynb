```python
#Popularity based reccomendation system
```

```python
ratings_with_name = ratings.merge(books, on = 'ISBN')
```

```python
num_rating_df = ratings_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_rating_df.rename(columns ={'Book-Rating': 'num_ratings'}, inplace = True)
num_rating_df
```

```python
avg_rating_df = ratings_with_name.groupby('Book-Title').mean()['Book-Rating'].reset_index()
avg_rating_df.rename(columns ={'Book-Rating': 'avg_ratings'}, inplace = True)
avg_rating_df
```

```python
popular_df = num_rating_df.merge(avg_rating_df, on = "Book-Title")
popular_df#merging the two dataframes on same column common to them book title
```

```python
popular_df = popular_df[popular_df['num_ratings']>=250].sort_values('avg_ratings', ascending = False).head(50) #i want the book having only reating greater than 250
```

```python
popular_df = popular_df.merge(books,on= 'Book-Title').drop_duplicates('Book-Title')[['Book-Title', 'Book-Author','Image-URL-M', 'num_ratings', 'avg_ratings']]
```

**Collaborative Filtering Based Recommender System**

```python
x = ratings_with_name.groupby('User-ID').count()['Book-Rating']>200
padhe_likhe_users = x[x].index
```

```python
filtered_rating = ratings_with_name[ratings_with_name['User-ID'].isin(padhe_likhe_users)]
```

```python
y = filtered_rating.groupby('Book-Title').count()['Book-Rating']>=50
famous_books =  y[y].index
```

```python
final_ratings = filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
```

```python
final_ratings.drop_duplicates()
```

```python
pt = final_ratings.pivot_table(index = 'Book-Title', columns = 'User-ID', values = 'Book-Rating' )
```

```python
pt.fillna(0,inplace = True)
```

```python
pt
```

```python
#Every book is a point in a 810 dimensional system
#Every book is now a vector in a 810 dimensional space
#And now we have to find similar vectors
#we have to find eucilidean distance of every book and with that only we can know which books are similar
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
#so basically i have found euclidean distance of 706 books with 706 books
cosine_similarity(pt).shape
```

```python
similarity_scores = cosine_similarity(pt)
```

```python
def recommend(book_name):
  index = np.where(pt.index == book_name)[0][0]
  similar_items = sorted(list(enumerate(similarity_scores[index])), key = lambda x:x[1],reverse = True)[1:6]#i dont want first one bcoz first one is with first one itself similarity score
  data =[]
  for i in similar_items:
    item = []
    temp_df = books[books['Book-Title'] == pt.index[i[0]]]
    item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
    item.extend(temp_df.drop_duplicates('Book-Title')['Book-Author'].values)
    item.extend(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values)

    data.append(item)
  return data

    #every time i will pick  an item and extract its thing, with this i get all these indexes
```

```python
recommend('1984')
```

```python
recommend('Message in a Bottle')
```

```python
recommend('The Da Vinci Code')
```
